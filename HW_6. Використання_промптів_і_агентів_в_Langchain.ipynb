Запустіть оновлений функціонал (промпт + модел) для пояснень про теми

Виведіть результати відпрацювання моделі на екран.

````python
from langchain_core.prompts import PromptTemplate
````

````python
prompt = PromptTemplate(
    input_variables=["topic"],
    template="""
    Поясни простою мовою тему: "{topic}".
    
    Відповідь обов'язково має містити:
    1) коротке визначення
    2) ключові переваги
    3) приклади поточних досліджень
    
    Обов'язково надавай відповідь на всі пункти з відображенням у вигляді списку!
    
    Обмеження:
    - скорочуй загальну довжину відповіді до 200 символів без різкого закінчення
    - максимально коротко
    """
)
````

````python
topics = [
    "Баєсівські методи в машинному навчанні",
    "Трансформери в машинному навчанні",
    "Explainable AI"
]

for topic in topics:
    print(f"Тема: {topic}")
    print(llm.invoke(prompt.format(topic=topic)))
    print("-" * 100)
````

````python
from langchain_openai import ChatOpenAI
from langchain_classic import hub
from langchain_classic.agents import load_tools, AgentExecutor, create_react_agent, Tool
from langchain_community.tools.tavily_search import TavilySearchResults
````

````python
# LLM 
llm = ChatOpenAI(model="gpt-4o", temperature=overal_temperature)

# Tool 
search_tool = TavilySearchResults(
    max_results=10,
    search_depth="advanced",
    include_answer=True,
    include_raw_content=False,
    include_images=False,
)
tools = [search_tool]

# prompt (сформований з допомогою GPT) 
TEMPLATE = """\
You are a scientific research assistant. Your only job is to find real academic publications.

RULES:
- Use ONLY the exact format below. No exceptions.
- After every Thought you MUST write Action and Action Input.
- Write Final Answer ONLY after you have real search results.
- Do NOT invent publications. Use only data from search results.

You have access to the following tools: {tools}

Use the following format:

Question: the input question you must answer
Thought: think about what to search
Action: tavily_search_results_json
Action Input: your search query as plain text
Observation: the result of the action
Thought: think about what you found, decide if you need another search
Action: tavily_search_results_json
Action Input: your next search query
Observation: the result of the action
Thought: I now have enough data to answer
Final Answer: write the list of 5 publications here

Begin!

Question: {input}
Thought:{agent_scratchpad}"""

prompt = PromptTemplate(
    input_variables=["input", "agent_scratchpad"],
    partial_variables={
        "tool_names": ", ".join(t.name for t in tools)
    },
    template=TEMPLATE
)

# Agent 
agent = create_react_agent(llm=llm, tools=tools, prompt=prompt)

agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=False,
    max_iterations=5,
    handle_parsing_errors=(
        "Format error! You MUST follow this exact pattern:\n"
        "Thought: ...\n"
        "Action: tavily_search_results_json\n"
        "Action Input: your query\n"
        "Do NOT write Final Answer without Action first."
    ),
    return_intermediate_steps=False,
)

# Функція 
def find_recent_publications(topic: str) -> str:
    result = agent_executor.invoke({
        "input": (
            f"Find 5 recent peer-reviewed papers about '{topic}' from 2025-2026 "
            f"For each: title, authors, year, 2-sentence description, URL."
        )
    })
    return result["output"]


if __name__ == "__main__":
    topic = "artificial intelligence"
    print(f"\n{'='*50}")
    print(f"Topic: {topic}")
    print(f"{'='*50}\n")

    output = find_recent_publications(topic)

    print("RESULTS:")
    print(output)
````

Створіть агента, який допомагає вирішувати задачі бізнес-аналітики. Агент має допомогти користувачу створити прогноз по продажам на наступний рік враховуючи рівень інфляції і погодні умови. Агент має вміти використовувати Python і ходити в інтернет аби отримати актуальні дані.

**Кроки:**
1. Налаштуйте агента, який працюватиме з аналітичними даними, заданими текстом. Користувач пише

```
Ми експортуємо апельсини з Бразилії. В 2022 експортували 200т, в 2023 - 190т, в 2024 - 210т, в 2025 - 220т. Зроби оцінку скільки ми зможемо експортувати апельсинів в 2026 враховуючи погодні умови в Бразилії і попит на апельсини в світі виходячи з економічної ситуації.
```

2. Створіть запит до агента, що містить чітке завдання – видати результат бізнес аналізу або написати, що він не може цього зробити і запит користувача (просто може бути все одним повідомлленням).

3. Запустіть агента і проаналізуйте результати. Що можна покращити?


````python
from langchain_experimental.utilities import PythonREPL
from langchain_classic.agents import create_tool_calling_agent
````

````python
llm = ChatOpenAI(model="gpt-4o", temperature=0)

# Python tool
python_repl = PythonREPL()
python_tool = Tool(
    name="python_interpreter",
    func=python_repl.run,
    description=(
        "Executes Python code. Input must be a valid Python script as a string. "
        "Always use print() to output results. "
        "numpy and sklearn are available."
    )
)

# Tools
search_tool = TavilySearchResults(max_results=5, search_depth="advanced")
tools = [search_tool, python_tool]

# Промпт 
TEMPLATE = """\
You are a senior business analyst specializing in commodity exports, macroeconomics, and forecasting.

RULES:
- Always respond in Ukrainian.
- Use search tool to get real current data before making conclusions.
- Use python_repl for ALL calculations. Always add print() to see results.
- Do NOT invent numbers — base every claim on search results or provided data.

Your final report MUST contain:
1. Weather/climate outlook for Brazil orange regions (use search)
2. Demand for oranges in the world based on the economic situation(use search)
3. Three scenarios: optimistic / base / pessimistic. Take into account the results from the previous two steps (use python_repl with print())

Begin!

Question: {input}
Thought:{agent_scratchpad}"""

prompt = PromptTemplate(
    input_variables=["input", "agent_scratchpad"],
    partial_variables={
        "tool_names": ", ".join(t.name for t in tools)
    },
    template=TEMPLATE
)

# Agent
agent = create_tool_calling_agent(llm=llm, tools=tools, prompt=prompt)

agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=False,
    max_iterations=15,
    handle_parsing_errors=True,
)

# Запит
USER_QUERY = """
Ми експортуємо апельсини з Бразилії. В 2022 експортували 200т, в 2023 - 190т, в 2024 - 210т, в 2025 - 220т. 
Зроби оцінку скільки ми зможемо експортувати апельсинів в 2026 враховуючи:
- погодні умови в Бразилії
- попит на апельсини в світі виходячи з економічної ситуації.

Якщо не можеш зробити повний аналіз — чітко поясни чого не вистачає.
"""

if __name__ == "__main__":
    print("=" * 60)
    print("Business Analytics Agent — Orange Export Forecast")
    print("=" * 60)

    result = agent_executor.invoke({"input": USER_QUERY})

    print("FINAL REPORT:")
    print(result["output"])
````
